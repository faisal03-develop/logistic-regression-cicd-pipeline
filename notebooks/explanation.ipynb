# Credit Risk Classifier — Step-by-Step Explanation

This notebook walks through the math and implementation of the Logistic Regression model used in this project.

```python
import sys
sys.path.insert(0, '..')
import numpy as np
import matplotlib.pyplot as plt
from src.model import LogisticRegression
from src.preprocess import StandardScaler, preprocess_pipeline
```

## 1. The Sigmoid Function

Squashes any real number to (0, 1) — interpreted as probability of default.

```python
z = np.linspace(-10, 10, 200)
sigmoid = 1 / (1 + np.exp(-z))
plt.figure(figsize=(8,4))
plt.plot(z, sigmoid, 'b-', linewidth=2)
plt.axhline(0.5, color='r', linestyle='--', label='Decision boundary (0.5)')
plt.xlabel('z (linear combination)')
plt.ylabel('P(default)')
plt.title('Sigmoid Activation Function')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 2. Load Data & Train

```python
data = preprocess_pipeline('../data/dataset.csv')
model = LogisticRegression(learning_rate=0.05, n_iterations=1000, verbose=True)
model.fit(data['X_train'], data['y_train'])
```

## 3. Loss Curve

```python
plt.figure(figsize=(8,4))
plt.plot(model.loss_history, 'b-', linewidth=1.5)
plt.xlabel('Iteration')
plt.ylabel('Binary Cross-Entropy Loss')
plt.title('Training Loss Curve')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 4. Feature Weights (Interpretability)

```python
feature_names = data['feature_names']
weights = model.weights
plt.figure(figsize=(10,5))
colors = ['red' if w > 0 else 'green' for w in weights]
plt.barh(feature_names, weights, color=colors)
plt.xlabel('Weight (positive = increases default risk)')
plt.title('Logistic Regression Feature Weights')
plt.axvline(0, color='black', linewidth=0.8)
plt.tight_layout()
plt.show()
```
